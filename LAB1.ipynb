```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import linear_model

from sklearn.feature_extraction import DictVectorizer as DV
```

```python
# filepath of train Data 
filepath_train = '/Users/ishaansingh/CMPE-256/data/RA/train.csv'
# file path of test Data
filepath_test = '/Users/ishaansingh/CMPE-256/data/RA/test.csv'
```

##### Open Data From CSV 

```python

def open_csv_file(filepath):
    return pd.read_csv(filepath)  
```

```python
def collect_data(filepath):
    data = open_csv_file(filepath)
    rows = data.shape[0];columns = data.shape[1]
    return data, rows, columns 
```

```python
# Basic Data obtained from train 
data_train, rows_train, columns_train = collect_data(filepath_train) 
data_test, rows_test, columns_test = collect_data(filepath_test) 
data_test.columns
```

### Data Preprocessing 

```python
# Removing the target ('Salary') from train Dataset.
target = data_train["salary"]
train_data = data_train.drop('salary', axis=1)
# type(train_data.columns)
# target
total_data = data_train.append(data_test)

data_total = total_data.drop('salary', axis = 1)
data_total.columns
```

```python
list1 = ['id', 'worker_group_name', 'union_name','job_group', 'job_group_code','job','department_name', 'job_code']
data_total_v1 = data_total.drop(list1, axis = 1)
data_total_v2 = data_total_v1.fillna(0)
data_total_v2.head(5)






```

```python
data_total_v2.department_code = pd.factorize(data_total.department_code)[0]
# data_total_v2.job_group_code = pd.factorize(data_total_v2.job_group_code)[0]
# data_total_v2.job_code = pd.factorize(data_total_v2.job_code)[0]
data_total_v2.union_code = pd.factorize(data_total_v2.union_code)[0]

data_total_v2
```

```python
# Divide the data 
train_data  = data_total_v2[:30000]
test_data_v1 = data_total_v2[30000:]

```

```python
regr = linear_model.LinearRegression()
new_dataframe = regr.fit(train_data, target)
print('Coefficients: \n', regr.coef_)
```

```python
target_pred = regr.predict(test_data_v1)
```

```python
final_dataframe = pd.DataFrame(target_pred, index = data_test.id)
```
